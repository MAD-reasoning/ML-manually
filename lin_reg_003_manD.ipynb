```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=1000, n_features=100, n_informative=97, noise=15, random_state=101)
```

```python
X.shape
```

```python
y.shape
```

```python
X = np.concatenate((np.ones((len(X), 1)), X), axis=1)
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
coeff = np.dot(np.linalg.inv(np.dot(X_train.T, X_train)), np.dot(X_train.T, y_train))
coeff
```

```python
y_pred = np.dot(X_test, coeff)
y_pred
```

```python
y_test
```

```python
u = np.sum((y_test - y_pred) ** 2) / float(len(y_test))
v = np.sum((y_test - y_test.mean()) ** 2) / float(len(y_test))
r2 = 1 - (u/v)
```

```python
r2
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
# Estimate linear regression coefficients using stochastic gradient descent
def coefficients_sgd(X, y, coeff, l_rate, n_epoch):
    m, n = X.shape
    # coef = np.zeros(n)
    coef = coeff
    for epoch in range(n_epoch):
        sum_error = 0.0
        for i in range(m):
            yhat = np.dot(X[i], coef)
            error = y[i] - yhat
            sum_error += (error**2)
            for j in range(n):
                coef[j] += (l_rate * error * X[i, j])
        print('---> epoch=%d, lrate=%.3f, error=%.3f' % (epoch, l_rate, sum_error))
    return coef
```

```python
# Calculate coefficients
l_rate = 0.001
n_epoch = 10
```

```python
coef = coefficients_sgd(X_train, y_train, coeff, l_rate, n_epoch)
coef
```

```python
y_predd = np.dot(X_test, coef)
y_predd
```

```python
y_test
```

```python
u = np.sum((y_test - y_predd) ** 2) / float(len(y_test))
v = np.sum((y_test - y_test.mean()) ** 2) / float(len(y_test))
r2 = 1 - (u/v)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_predd)
```
