```python
from sklearn.datasets import load_iris
```

```python
# Dataset
dataset = load_iris()
```

```python
# X
X = dataset.data
X.shape
```

```python
# y
y = dataset.target
y.shape
```

```python
# Splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
a,b = X_train.shape
a,b
```

```python
elem = []
for i in range(a):
    if y_train[i] not in elem:
        elem.append(y_train[i])
elem.sort()        
elem
```

```python
def mean(numbers):
    return sum(numbers)/float(len(numbers))

def stdev(numbers):
    avg = mean(numbers)
    variance = sum([pow(x-avg,2) for x in numbers])/float(len(numbers)-1)
    return math.sqrt(variance)
```

```python
summr = {}
for j in elem:
    elemv = []
    lister = []
    for i in range(a):
        if y_train[i] == elem[j]:
            elemv.append(X_train[i])
    for k in range(b):
        lists = []
        for l in range(len(elemv)):
            lists.append(elemv[l][k])
        lister.append((mean(lists), stdev(lists)))
    summr[elem[j]] = (lister)
summr
```

```python
def calculateProbability(x, mean, stdev):
    exponent = math.exp(-(math.pow(x-mean,2)/(2*math.pow(stdev,2))))
    return (1 / (math.sqrt(2*math.pi) * stdev)) * exponent
```

```python
c, d = X_test.shape
c, d
```

```python
lister = []
for i in range(c):
    lists = []
    for k in elem:
        elemv = 1
        for j in range(d):
            elemv *= (calculateProbability(X_test[i][j], summr[elem[k]][j][0], summr[elem[k]][j][1]))
        lists.append(elemv)
    lister.append(lists)
```

```python
lister
```

```python
y_pred = []
for i in range(c):
    num, val = 0, None
    for j in range(len(elem)):
        if lister[i][j] > num:
            num = lister[i][j]
            val = elem[j]
    y_pred.append(val)
```

```python
np.array(y_pred)
```

```python
y_test
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
